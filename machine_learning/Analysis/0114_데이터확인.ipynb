```python
import pandas as pd

def calculate_shipping_duration(df):
    # 'Order Date'와 'Ship Date' 열을 datetime 형식으로 변환
    df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
    df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
    
    # 두 날짜의 차이를 계산하여 새로운 열 'Shipping Duration'에 저장
    df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
    
    return df
```

```python
def extract_coordinates(point):
    # 'POINT (' 와 ')' 제거 후 공백으로 분리
    coords = point.replace('POINT (', '').replace(')', '').split()
    return pd.Series({'lng': float(coords[0]), 'lat': float(coords[1])})

# 함수 적용하여 새로운 열 생성
location[['lng', 'lat']] = location['centroid'].apply(extract_coordinates)
```

### Postal Code 결측치 처리
1. Order ID
2. Customer ID
3. Drop

> Order ID가 같은 것으로 결측치 대체

```python
# Order ID를 기준으로 Postal Code의 딕셔너리 생성 (결측치 제외)
order_id_to_postal_code = data.dropna(subset=['Postal Code']).groupby('Order ID')['Postal Code'].first().to_dict()

# 결측치를 딕셔너리를 이용해 채우기
data['Postal Code'] = data.apply(
    lambda row: order_id_to_postal_code.get(row['Order ID'], row['Postal Code']) 
    if pd.isnull(row['Postal Code']) else row['Postal Code'],
    axis=1
)
```

```python
data.info()
```

> Order ID 기준으로는 결측치가 채워지지 않았다.

> Customer ID가 같은 것으로 결측치 대체

```python
# Customer ID를 기준으로 Postal Code의 딕셔너리 생성 (결측치 제외)
order_id_to_postal_code = data.dropna(subset=['Postal Code']).groupby('Customer ID')['Postal Code'].first().to_dict()

# 결측치를 딕셔너리를 이용해 채우기
data['Postal Code'] = data.apply(
    lambda row: order_id_to_postal_code.get(row['Customer ID'], row['Postal Code']) 
    if pd.isnull(row['Postal Code']) else row['Postal Code'],
    axis=1
)
```

```python
data.info()
```

> Customer ID가 같은 기준으로 결측치가 모두 채워졌다.

```python
data['Postal Code'] = data['Postal Code'].astype('int64')
data.info()
```
